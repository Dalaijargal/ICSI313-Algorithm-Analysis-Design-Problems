```python
import networkit as nk
import matplotlib.pyplot as plt
import numpy as np
```

```python
pgp = nk.graphio.METISGraphReader().read("/home/temuujin/PGPgiantcompo.graph")

print("Number of nodes:", pgp.numberOfNodes())
print("Number of edges:", pgp.numberOfEdges())
```

```python
dd = sorted(nk.centrality.DegreeCentrality(pgp).run().scores(), reverse=True)
x = np.arange(0.07, 6, 0.1);
y = 1/(x**2)

plt.xlabel("degree")
plt.ylabel("frequency")
plt.title("PGPgiantcompo")
plt.plot(dd)
plt.show()

plt.xlabel("x")
plt.ylabel("y")
plt.title("y = x^(-2)")
plt.plot(x, y)
plt.show()
```
